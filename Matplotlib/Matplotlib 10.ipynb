# Box Plot / Box Chart

Import Modules 

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print (matplotlib.__version__)
print (np.__version__)
```

## Sample Dataset

```python
np.random.seed(2)
data=np.random.normal(loc=100,scale=10,size=200)
data
```

## Simple Box Plot

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Notched Box Plot 

```python
plt.boxplot(data,labels=['Data'],notch=True)

plt.title('Notched Box Plot ')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show ()
```

## Pengaturan Marker untuk Outlier

```python
plt.boxplot(data,
            labels=['Data'],
            showfliers=True,
           flierprops ={'markerfacecolor':'red', 'marker':'d'})

plt.title('Pengaturan Marker ')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show ()
```

## Horisontal Box Plot 

```python
plt.boxplot(data,labels=['Data'],vert=False)

plt.title('Pengaturan Marker ')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show ()
```

# Multiple Box Plot


```python
np.random.seed(10)
data1 =np.random.normal(100,10,200)
data2 =np.random.normal(80,30,200)
data3 =np.random.normal(90,20,200)
data4 =np.random.normal(70,25,200)

data =[data1,data2,data3,data4]
labels=['Data 1','Data2','Data3', 'Data4']
```

```python
plt.boxplot(data,labels=labels)

plt.title('Multiple Box Plot ')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show ()
```
